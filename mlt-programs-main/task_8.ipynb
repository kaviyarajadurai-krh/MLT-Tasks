```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import datasets

# Read Dataset
df = datasets.load_iris()
x = df.data
y = df.target

print("Feature Data:\n", x)
print("Target Labels:\n", y)

# Try with k = 5 initially
kmeans5 = KMeans(n_clusters=5, random_state=42)
y_kmeans5 = kmeans5.fit_predict(x)

print("Cluster labels for k=5:\n", y_kmeans5)
print("Cluster centers for k=5:\n", kmeans5.cluster_centers_)

# Find optimal number of clusters using Elbow Method
Error = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(x)
    Error.append(kmeans.inertia_)

# Plot Elbow graph
plt.plot(range(1, 11), Error, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia (Error)')
plt.show()

# Now try with k = 3
kmeans3 = KMeans(n_clusters=3, random_state=42)
y_kmeans3 = kmeans3.fit_predict(x)

print("Cluster labels for k=3:\n", y_kmeans3)
print("Cluster centers for k=3:\n", kmeans3.cluster_centers_)
```
