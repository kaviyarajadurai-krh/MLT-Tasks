```python
# Gaussian Mixture Model Classifier Example

import numpy as np
from sklearn.datasets import load_iris
from sklearn.mixture import GaussianMixture
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from scipy.stats import mode

# Load dataset
data = load_iris()
X = data.data
y = data.target

# Split dataset into training and testing
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Number of classes
n_classes = len(np.unique(y))

# Train Gaussian Mixture Model
gmm = GaussianMixture(n_components=n_classes, random_state=42)
gmm.fit(X_train)

# Predict clusters
train_clusters = gmm.predict(X_train)
test_clusters = gmm.predict(X_test)

# Map clusters to actual labels
labels = np.zeros(n_classes)

for i in range(n_classes):
    mask = (train_clusters == i)
    labels[i] = mode(y_train[mask])[0]

labels = labels.astype(int)

# Convert cluster predictions to class predictions
y_pred = np.zeros_like(test_clusters)

for i in range(len(test_clusters)):
    y_pred[i] = labels[test_clusters[i]]

# Compute accuracy
accuracy = accuracy_score(y_test, y_pred)

print("Predicted Labels:", y_pred)
print("Actual Labels:", y_test)
print("Accuracy of GMM Classifier:", accuracy)
```
